```python
#Drag coefficient and frontal area scraper
res = requests.get('http://tech-racingcars.wikidot.com/aerodynamics')
soup = bs4.BeautifulSoup(res.text,'html.parser')
table1 = soup.find('table',class_ = 'wiki-content-table' )

rows = table1.find_all('tr')
for row in rows[1:-1]:
    cells = row.find_all(['td','th'])
    cells_text = list([cell.getText() for cell in cells])
    print(cells_text)
```
